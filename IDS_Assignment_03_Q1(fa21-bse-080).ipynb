```python
# 16-April-2024
# CSC461 – Assignment3 – IDS – Graph Analysis
# Abdul ul Rehman Nasar
# FA21-BSE-080
# Using the networkx library in Python to create an undirected graph described by the adjacency matrix given in the assignment and then checking if there any loop present in the graph or not and if present then at which point.


#import libraries for creating and printing grpahs in Python

import networkx as nx
import matplotlib.pyplot as plt
```

```python
#create a simple (non-directed) graph

sG = nx.Graph()
```

```python
#add nodes and edges to the simple graph and display it

sG.add_node("A")
sG.add_node("B")
sG.add_node("C")
sG.add_node("D")
sG.add_node("E")
sG.add_node("F")
sG.add_node("G")
sG.add_edge("A", "C")
sG.add_edge("A", "E")
sG.add_edge("A", "G")
sG.add_edge("B", "B")
sG.add_edge("B", "F")
sG.add_edge("C", "A")
sG.add_edge("C", "F")
sG.add_edge("C", "G")
sG.add_edge("D", "E")
sG.add_edge("E", "A")
sG.add_edge("E", "D")
sG.add_edge("F", "B")
sG.add_edge("F", "C")
sG.add_edge("F", "G")
sG.add_edge("G", "A")
sG.add_edge("G", "C")
sG.add_edge("G", "F")
```

```python
# Draw the graph

nx.draw(sG, with_labels = True)
```

```python
# Check for loops

loops = list(nx.simple_cycles(sG))
if loops:
    print("There is a loop in the graph.")
    print("The loop is at the following nodes:", loops[0])
else:
    print("There are no loops in the graph.")
```
