```python
# 16-April-2024
# CSC461 – Assignment3 – IDS – Graph Analysis
# Abdul ul Rehman Nasar
# FA21-BSE-080
# Assigning the random weights to the edges of the graph created in question 01 and then finding the shortest path between nodes A and B using Dijkstra's algorithm
#import libraries for creating and printing grpahs in Python

import networkx as nx
import matplotlib.pyplot as plt
```

```python
#create a simple (non-directed) graph

sG = nx.Graph()
```

```python
#add nodes and edges to the simple graph and display it

sG.add_node("A")
sG.add_node("B")
sG.add_node("C")
sG.add_node("D")
sG.add_node("E")
sG.add_node("F")
sG.add_node("G")
sG.add_edge("A", "C", weight= 3)
sG.add_edge("A", "E", weight= 5)
sG.add_edge("A", "G", weight= 7)
sG.add_edge("B", "B", weight= 6)
sG.add_edge("B", "F", weight= 4)
sG.add_edge("C", "A", weight= 7)
sG.add_edge("C", "F", weight= 2)
sG.add_edge("C", "G", weight= 1)
sG.add_edge("D", "E", weight= 9)
sG.add_edge("E", "A", weight= 4)
sG.add_edge("E", "D", weight= 6)
sG.add_edge("F", "B", weight= 7)
sG.add_edge("F", "C", weight= 5)
sG.add_edge("F", "G", weight= 3)
sG.add_edge("G", "A", weight= 7)
sG.add_edge("G", "C", weight= 8)
sG.add_edge("G", "F", weight= 2)
```

```python
# Draw the graph

nx.draw(sG, with_labels = True)
```

```python
# Find the shortest path using Dijkstra's algorithm

shortest_path = nx.dijkstra_path(sG, source='A', target='B')
shortest_path_length = nx.dijkstra_path_length(sG, source='A', target='B')

print("Shortest Path:", shortest_path)
print("Length of Shortest Path:", shortest_path_length)
```
